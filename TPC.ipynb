```python
from operator import truediv
import random


print("(1) Criar Lista")
print("(2) Ler Lista")
print("(3) Soma")
print("(4) Média")
print("(5) Maior")
print("(6) Menor")
print("(7) estaOrdenada por ordem crescente")
print("(8) estaOrdenada por ordem decrescente")
print("(9) Procurar um elemento")
print("(0) Sair")
n = 10
ListaExist = False
res = []
max = 0
min = 101
soma = 0
média = 0
oldfound = 0
while n != 0:
    n = int(input())



    if n == 1:
        print("insira tamanho da lista")
        t = int(input())
        max = 0
        min = 101
        soma = 0
        média = 0
        ListaExist = True
        res = []
        num = 1
        l = t
        rand = 0
        while num <= t:
            rand = random.randint(1, 100)
            res.append(rand)
            num = num + 1
            soma = soma + rand
            média = soma/l
            if (rand > max):
                max = rand
            if (rand < min):
                min = rand
    if n == 2:
        if ListaExist == True:
            print(res)
        if ListaExist != True:
            print("Ainda não criou uma lista")
    if n == 3:
        if ListaExist == True:
            print(soma)
        if ListaExist != True:
            print("Ainda não criou uma lista")
    if n == 4:
        if ListaExist == True:
            print(média)
        if ListaExist != True:
            print("Ainda não criou uma lista")
    if n == 5:
        if ListaExist == True:
            print(max)
        if ListaExist != True:
            print("Ainda não criou uma lista")
    if n == 6:
        if ListaExist == True:
            print(min)
        if ListaExist != True:
            print("Ainda não criou uma lista")
    if n == 7:
        if ListaExist == True:
            res.sort()
        if ListaExist != True:
            print("Ainda não criou uma lista")
    if n == 8:
        if ListaExist == True:
            res.sort(reverse=True)
        if ListaExist != True:
            print("Ainda não criou uma lista")
    if n == 9:
        print("Que número quer procurar?")
        search = int(input())
        for v in res:
            if v == search:
                found = res.index(v)
                if found == oldfound:
                    oldfound = found
                    found = found + 1
                if found != oldfound:
                    oldfound = found
                print("Esse número foi encontrado em ", found)
            

        
    if n == 0:
        print("Fim do programa")
    



```
